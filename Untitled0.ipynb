```python
#K Means Clustering
```

```python
data.columns
```

```python
x=data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
k=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=0)
  kmeans.fit(x)
  k.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),k)
plt.show()
```

```python
model=KMeans(n_clusters=5 , init="k-means++",random_state=0)
y_kemsn=model.fit_predict(x)
```

```python
plt.figure(1, figsize=(15,8))
plt.scatter(x[y_kemsn==0,0],x[y_kemsn==0,1],s=100,c='magenta',label='Cluster 1')
plt.scatter(x[y_kemsn==1,0],x[y_kemsn==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(x[y_kemsn==2,0],x[y_kemsn==2,1],s=100,c='cyan',label='Cluster 3')
plt.scatter(x[y_kemsn==3,0],x[y_kemsn==3,1],s=100,c='green',label='Cluster 4')
plt.scatter(x[y_kemsn==4,0],x[y_kemsn==4,1],s=100,c='red',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c="black",label="Centroids")
plt.title('K Means Clustering Algorithm')
plt.xlael('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv('Mall_Customers.csv')

# Select features for clustering (Age, Annual Income, Spending Score)
X = data.iloc[:, [2, 3, 4]].values

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Add cluster labels to the original data
data['Cluster'] = labels

# Visualize the clusters (scatter plot)
plt.scatter(X_scaled[labels == 0, 0], X_scaled[labels == 0, 1], s=50, c='red', label='Cluster 0')
plt.scatter(X_scaled[labels == 1, 0], X_scaled[labels == 1, 1], s=50, c='blue', label='Cluster 1')
plt.scatter(X_scaled[labels == -1, 0], X_scaled[labels == -1, 1], s=50, c='green', label='Noise')
plt.xlabel(' Age')
plt.ylabel(' Annual Income')
plt.title('DBSCAN Clustering')
plt.legend()
plt.show()

print(data[['CustomerID', 'Cluster']])
```
